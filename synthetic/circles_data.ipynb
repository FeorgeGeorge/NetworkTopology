```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def circle(num_points, center, radius):
    x_0, y_0 = center

    l = np.linspace(0, 2*np.pi, num=num_points)
    x = x_0 + radius*(np.cos(l)) + np.random.normal(size=num_points)
    y = y_0 + radius*(np.sin(l)) + np.random.normal(size=num_points)
    return x, y
```

```python
x_green1, y_green1 = circle(100, (0, 0), 10)

x_blue1, y_blue1 = circle(100, (20, 20), 10)
x_green2, y_green2 = circle(100, (20, 20), 4)

x_blue2, y_blue2 = circle(100, (0,0), 4)
x_blue3, y_blue3 = circle(100, (20, -10), 4)
x_green3, y_green3 = circle(100, (-10, 20), 4)
```

```python
blue_x = np.concatenate((x_blue1, x_blue2, x_blue3)) 
blue_y = np.concatenate((y_blue1, y_blue2, y_blue3))

green_x = np.concatenate((x_green1, x_green2, x_green3))
green_y = np.concatenate((y_green1, y_green2, y_green3))

data_blue = [np.array([blue_x[i], blue_y[i]]) for i in range(len(blue_x))]
data_green = [np.array([green_x[i], green_y[i]]) for i in range(len(blue_x))]
```

```python
data = np.array(data_blue + data_green)
target = [0] * len(data_blue) + [1] * len(data_green)
```

```python
def draw_solid_figure(data):    
    plt.scatter(data[0], data[1], c = data["target"])
```

```python
data_normalised = (data - data.mean()) / (data.std())

data_total = pd.DataFrame(data_normalised)
data_total["target"] = target
```

```python
draw_solid_figure(data_total)
```

```python
pd.DataFrame(data_total).to_csv("data_new_circles.csv", encoding="utf-8", index=False)
```
